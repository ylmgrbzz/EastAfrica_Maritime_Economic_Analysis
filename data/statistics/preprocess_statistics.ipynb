```python
import pandas as pd
import json
```

```python
# Define a list of countries and their corresponding codes
countries = ['BURUNDI', 'CONGO', 'KENYA', 'RWANDA', 'SOMALIA', 'TANZANIA', 'UGANDA']
country_codes = ['BDI', 'COD', 'KEN', 'RWA', 'SOM', 'TZA', 'UGA']
```

```python
# Create an empty list to store DataFrames

dfs = []
for country in countries:
    df = pd.read_excel('{}.xlsx'.format(country))
    
    dfs.append(df)
```

```python
# Concatenate all DataFrames into a single DataFrame
df = pd.concat(dfs, ignore_index=True)
```

```python
# Save the combined DataFrame to an Excel file
df.to_excel('all.xlsx', index=False)
```

```python
# Extract unique series names from the 'Series Name' column
series_names = df['Series Name'].unique()
```

```python
# Define new columns for the final DataFrame
new_columns = ['Country Name', 'Country Code'] + list(series_names) 
```

```python
# Generate a range of years from 2005 to 2022
dates = range(2005, 2023)
```

```python
# Read additional Excel files for vessel calls and container traffic
vessel_calls = pd.read_excel('../2005 TO 2022_VESSEL CALLS.xlsx')
traffic_df = pd.read_excel('../2005 TO 2022_CONTAINER TRAFFIC TEU.xlsx')
```

```python
# Create a list to store new rows for the final DataFrame

new_rows = []

 # Loop through each year and country code to populate the new rows
           
for year in dates:
    new_row = {
        'Year': year,
    }
    
    for code in country_codes:
        # filter by country
        filtered = df[df['Country Code'] == code]
        new_row[code + '-' +series_names[0]] = filtered[filtered['Series Name'] == series_names[0]][year].iloc[0]
        new_row[code + '-' +series_names[1]] = filtered[filtered['Series Name'] == series_names[1]][year].iloc[0]
        new_row[code + '-' +series_names[2]] = filtered[filtered['Series Name'] == series_names[2]][year].iloc[0]
        new_row[code + '-' +series_names[3]] = filtered[filtered['Series Name'] == series_names[3]][year].iloc[0]
        new_row[code + '-' +series_names[4]] = filtered[filtered['Series Name'] == series_names[4]][year].iloc[0]
        new_row[code + '-' +series_names[5]] = filtered[filtered['Series Name'] == series_names[5]][year].iloc[0]
        new_row[code + '-' +series_names[6]] = filtered[filtered['Series Name'] == series_names[6]][year].iloc[0]
        new_row[code + '-' +series_names[7]] = filtered[filtered['Series Name'] == series_names[7]][year].iloc[0]
        new_row[code + '-' +series_names[8]] = filtered[filtered['Series Name'] == series_names[8]][year].iloc[0]
        new_row[code + '-' +series_names[9]] = filtered[filtered['Series Name'] == series_names[9]][year].iloc[0]
        new_row[code + '-' +series_names[10]] = filtered[filtered['Series Name'] == series_names[10]][year].iloc[0]
        new_row[code + '-' +series_names[11]] = filtered[filtered['Series Name'] == series_names[11]][year].iloc[0]
        new_row[code + '-' +series_names[12]] = filtered[filtered['Series Name'] == series_names[12]][year].iloc[0]
        new_row[code + '-' +series_names[13]] = filtered[filtered['Series Name'] == series_names[13]][year].iloc[0]
        new_row[code + '-' +series_names[14]] = filtered[filtered['Series Name'] == series_names[14]][year].iloc[0]
        new_row[code + '-' +series_names[15]] = filtered[filtered['Series Name'] == series_names[15]][year].iloc[0]
        new_row[code + '-' +series_names[16]] = filtered[filtered['Series Name'] == series_names[16]][year].iloc[0]
        new_row[code + '-' +series_names[17]] = filtered[filtered['Series Name'] == series_names[17]][year].iloc[0]
        new_row[code + '-' +series_names[18]] = filtered[filtered['Series Name'] == series_names[18]][year].iloc[0]

    new_row['vessel_calls'] = vessel_calls[year].iloc[0]
    new_row['container_port_traffic'] = traffic_df[year].iloc[0]
    new_rows.append(new_row)
        
```

```python
new_df = pd.DataFrame(new_rows)
```

```python
# Create separate DataFrames for vessel calls and container traffic

vessel_calls_dataset = new_df.drop(['container_port_traffic'], axis=1)
vessel_calls_dataset.to_excel('vessel_calls_dataset.xlsx', index=False)
```

```python
# Save the DataFrames to Excel files

traffic_dataset = new_df.drop(['vessel_calls'], axis=1)
traffic_dataset.to_excel('traffic_dataset.xlsx', index=False)
```
