```python
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd

# Loading and preparing the dataset (Assumes file path and content based on assumption)
file_path = 'data.xlsx'  # Enter your actual file path here
df = pd.read_excel(file_path)

# Selecting the target variable
y = df['container_port_traffic']

# Building and training the ARIMA model
model_arima = ARIMA(y, order=(1, 1, 1))
model_arima_fit = model_arima.fit()

# Making predictions for the next 20 years
forecast_arima = model_arima_fit.forecast(steps=20)

# For example, simulating predicted values and confidence intervals (add your actual prediction results here)

# Converting predicted values and confidence intervals to a DataFrame
forecast_df_example = pd.DataFrame({
    'Year': range(2023, 2043),
    'Predicted container_port_traffic': forecast_arima,
})

forecast_df_example
```

```python
import matplotlib.pyplot as plt

# Creating a DataFrame containing actual values
actual_df = pd.DataFrame({
    'Year': df['Year'].tail(20),  # Last 20 years for predictions
    'Actual container_port_traffic': df['container_port_traffic'].tail(20),
})

# Merging DataFrames containing predicted and actual values
result_df = pd.merge(actual_df, forecast_df_example, on='Year', how='left')

# Creating a plot for visual analysis
plt.figure(figsize=(12, 6))
plt.plot(result_df['Year'], result_df['Actual container_port_traffic'], label='Actual', marker='o')
plt.plot(result_df['Year'], result_df['Predicted container_port_traffic'], label='Predicted', marker='o')
plt.fill_between(result_df['Year'], result_df['Predicted container_port_traffic'].min(), result_df['Predicted container_port_traffic'].max(), color='gray', alpha=0.3, label='Confidence Interval')
plt.title('Actual vs Predicted Container Port Traffic')
plt.xlabel('Year')
plt.ylabel('Container Port Traffic (TEU: 20 foot equivalent units)')
plt.legend()
plt.show()
```

```python
# Plotting Population Growth over the years
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['BDI-Population growth (annual %)'], label='Population Growth', marker='o', linestyle='--', color='green')
plt.title('Population Growth Over the Years')
plt.xlabel('Year')
plt.ylabel('Population Growth (annual %)')
plt.legend()
plt.show()

# Plotting GNI per capita over the years
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['BDI-GNI per capita, Atlas method (current US$)'], label='GNI per capita', marker='o', linestyle='--', color='blue')
plt.title('GNI per Capita Over the Years')
plt.xlabel('Year')
plt.ylabel('GNI per Capita (current US$)')
plt.legend()
plt.show()

# Plotting Inflation over the years
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['BDI-Inflation, GDP deflator (annual %)'], label='Inflation', marker='o', linestyle='--', color='red')
plt.title('Inflation Over the Years')
plt.xlabel('Year')
plt.ylabel('Inflation (annual %)')
plt.legend()
plt.show()
```

```python
# Plotting GDP growth over the years
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['BDI-GDP growth (annual %)'], label='GDP Growth', marker='o', linestyle='--', color='purple')
plt.title('GDP Growth Over the Years')
plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.legend()
plt.show()

# Plotting Exports and Imports as a percentage of GDP
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['BDI-Exports of goods and services (% of GDP)'], label='Exports (% of GDP)', marker='o', linestyle='--', color='orange')
plt.plot(df['Year'], df['BDI-Imports of goods and services (% of GDP)'], label='Imports (% of GDP)', marker='o', linestyle='--', color='brown')
plt.title('Exports and Imports as Percentage of GDP Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage of GDP')
plt.legend()
plt.show()

```
